```python
import os
import glob
import pandas as pd


root_dir = "/home/ndecaux/NAS_EMPENN/share/projects/actidep/bids/derivatives/bundle_seg"

subjects = glob.glob(os.path.join(root_dir, "sub-*"))

info_list = []
for subject in subjects:
    print(subject)
    # Get the list of all files in the subject directory
    files = glob.glob(os.path.join(subject, 'tracto', "*.trk"))
    bundle_names = [f.split('bundle-')[1].split('_desc')[0] for f in files]
    info_dict = {f:True for f in bundle_names}
    info_dict['subject'] = subject.split('/')[-1]
    info_list.append(info_dict)

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(info_list)

#Set subject as index
df.set_index('subject', inplace=True)
# Save the DataFrame to a xlsx file
output_file = os.path.join(root_dir, "bundle_seg.xlsx")
# Apply a style to highlight NaN values with a red background
def highlight_nan(val):
    color = 'red' if pd.isna(val) else ''
    return f'background-color: {color}'

styled_df = df.style.applymap(highlight_nan)
styled_df.to_excel(output_file, index=True, engine='openpyxl')
print(f"Excel file saved to {output_file}")
```
