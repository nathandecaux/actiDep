```python
from os.path import join as opj
root_dir = '/home/ndecaux/NAS_EMPENN/share/projects/HCP105_Zenodo_NewTrkFormat/inGroupe1Space'

atlas_path = opj(root_dir, 'BundleSegMultiAtlas')

#Get all subjects that are in root_dir (6-digit folder names)
import os
import glob

subjects = [f for f in os.listdir(root_dir) if os.path.isdir(opj(root_dir, f)) and len(f) == 6]
subjects.sort()
print('Found %d subjects in %s' % (len(subjects), root_dir))

# --atlas_path                  Path of the folder containing atlas
#                                     ├── atlas
#                                     │   └── <subject_id>
#                                     ├── config_fss_1.json
#                                     └── mni_masked.nii.gz
#Create symbolic links to the atlas
import shutil
import pathlib
#First create the atlas folder
atlas_dir = opj(root_dir, 'BundleSegMultiAtlas', 'atlas')
if not os.path.exists(atlas_dir):
    os.makedirs(atlas_dir)
    
#Then create folders for each subject and symlink only the .trk files
for subject in subjects:
    #Create the subject folder
    subject_dir = opj(atlas_dir, subject)
    if not os.path.exists(subject_dir):
        os.makedirs(subject_dir)
        print(f"Created directory for {subject}")
    
    #Find all .trk files in the original subject directory
    original_subject_dir = opj(root_dir, subject)
    trk_files = glob.glob(opj(original_subject_dir, "*.trk"))
    
    if trk_files:
        #Create symbolic links for each .trk file
        for trk_file in trk_files:
            trk_filename = os.path.basename(trk_file)
            src = trk_file
            dst = opj(subject_dir, trk_filename)
            
            if not os.path.exists(dst):
                os.symlink(src, dst)
                print(f"Created symbolic link for {trk_filename} in {subject}")
            else:
                print(f"Symbolic link already exists for {trk_filename} in {subject}")
    else:
        print(f"No .trk files found in {original_subject_dir}")

#Create a symlink from /home/ndecaux/NAS_EMPENN/share/projects/HCP105_Zenodo_NewTrkFormat/inGroupe1Space/Atlas/average_anat.nii.gz to /home/ndecaux/NAS_EMPENN/share/projects/HCP105_Zenodo_NewTrkFormat/inGroupe1Space/BundleSegMultiAtlas/average_anat.nii.gz
src = opj(root_dir, 'Atlas', 'average_anat.nii.gz')
dst = opj(root_dir, 'BundleSegMultiAtlas', 'average_anat.nii.gz')
if not os.path.exists(dst):
    os.symlink(src, dst)
    print('Created symbolic link for average_anat.nii.gz')
else:
    print('Symbolic link already exists for average_anat.nii.gz')
```
