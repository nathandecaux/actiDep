```python
import os
import sys
import glob
import pathlib
import shutil
root_dir = "/home/ndecaux/NAS_EMPENN/share/projects/actidep/BundleSegResults/missing_bundles/"
base_dir = os.path.dirname(root_dir) # Get the parent directory of root_dir

sub_list = glob.glob(os.path.join(root_dir, "sub-*"))
sub_list.sort()

#For each subject folder, get the folder where files that ends with _cleaned.trk are stored

good_dir={}
for sub in sub_list:
    #Get the list of all folders that contain files ending with _cleaned.trk
    print(os.path.join(sub, "*", "*_cleaned.trk"))
    folder_list = glob.glob(os.path.join(sub, "*","*", "*", "*_cleaned.trk"))
    folder_list.sort()
    sub_id = os.path.basename(sub)

    #For each folder, get the name of the folder
    if os.path.exists(
            os.path.join(base_dir, "bids", "derivatives", "bundle_seg",
                         sub_id)):
        print('Skipping subject', sub_id)
        continue
    for folder in folder_list:
        #Get the name of the folder
        folder_name = os.path.basename(os.path.dirname(folder))
        #Get the name of the file
        file_name = os.path.basename(folder)
        #Get the parent folder
        parent_folder = os.path.basename(
            os.path.dirname(os.path.dirname(folder)))
        #Get the name of the subject
        target_dir = os.path.join(sub, parent_folder, folder_name)
        good_dir[sub_id] = target_dir

        #Extract in the file what's between summed_ and _cleaned
        bundle_name = file_name.split("_cleaned")[0].split(
            "summed_")[1].replace("_", "")

        #The target name shoudl be sub-<sub_id>_bundle-<bundle_name>_desc-cleaned_tracto.trk
        target_name = f"{sub_id}_bundle-{bundle_name}_desc-cleaned_tracto.trk"

        #Target path is /home/ndecaux/NAS_EMPENN/share/projects/actidep/bids/derivatives/bundle_seg/sub-<sub_id>/tracto/sub-<sub_id>_bundle-<bundle_name>_desc-cleaned_tracto.trk

        target_path = os.path.join(base_dir, "bids", "derivatives",
                                   "bundle_seg", sub_id, "tracto", target_name)
        #Create the target directory if it doesn't exist
        os.makedirs(os.path.dirname(target_path), exist_ok=True)
        #Copy the file to the target directory

        shutil.copy(folder, target_path)

        print(f"Copying {folder} to {target_path}")

        #Print the name of the subject, folder and file
        # print(f"{sub_name} {folder_name} {file_name}")
        #

print(good_dir)
```
