```python
from actiDep.data.loader import Subject
from tractviewer import TractViewer

sub = Subject('03008', db_root ='/home/ndecaux/NAS_EMPENN/share/projects/actidep/bids/')
original_bundles = sub.get(pipeline='bundle_seg',extension='trk')
autocal_bundles= sub.get(pipeline='bundle_seg_autocalib',extension='trk')
```

```python
print(len(original_bundles), len(autocal_bundles))
    # vis.record_rotation("rotation.mp4", n_frames=240, step=1.5, quality=10)

for b in autocal_bundles:
    vis = TractViewer(background="white", off_screen=True)
    bundle_name=b.get_entities()['bundle']
    print(bundle_name)
    vis.add_dataset(b.path, color='green', opacity=0.5)
    b_original = [bo for bo in original_bundles if bo.get_entities()['bundle'] == bundle_name]
    if len(b_original) == 1:
        vis.add_dataset(b_original[0].path, name=bundle_name + '_original', color='red', opacity=0.5)
    vis.record_rotation("vid_autocalib/rotation_{}.mp4".format(bundle_name), n_frames=240, step=1.5, quality=10)
```

```python
vis = TractViewer(background="white", off_screen=True)

for b in autocal_bundles:
    bundle_name=b.get_entities()['bundle']
    print(bundle_name)
    vis.add_dataset(b.path, color='green', opacity=0.5)
    b_original = [bo for bo in original_bundles if bo.get_entities()['bundle'] == bundle_name]
    if len(b_original) == 1:
        vis.add_dataset(b_original[0].path, name=bundle_name + '_original', color='red', opacity=0.5)

vis.record_rotation("vid_autocalib/rotation_whole_brain.mp4".format(bundle_name), n_frames=240, step=1.5, quality=10)
```
